```python
#Preguntan.°1: ¿Cuántos pedidos únicos hay en el archivo orders.csv?
from os.path import exists
import csv
dataorders="/content/sample_data/orders.csv"
if not exists(dataorders):
  raise SystemExit("You should run the first code cell and download the dataset files!" )

with open(dataorders, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)
  unique_orders = set()
  for row in csvreader:
    unique_orders.add(row[1])
num_unique_orders = len(unique_orders)
print(f"There are {num_unique_orders} unique orders in the orders.csv file.")
```
